```python
s = "hello world"

s.upper()    # "HELLO WORLD"
s.lower()    # "hello world"
s.title()    # "Hello World"
s.capitalize()  # "Hello world"
#note the last output as we didn't include any print commands
```

```python
s = "  hello  "
print(s)
s.strip()   # "hello"
s.lstrip()  # "hello  "
s.rstrip()  # "  hello"
```
